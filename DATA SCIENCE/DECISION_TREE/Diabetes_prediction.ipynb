### Diabetes Prediction

```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd
df_diabetes=pd.DataFrame(dataset.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
x=df_diabetes
y=dataset['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.head()
```

```python
x_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
sns.heatmap(x_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

##Hyperparameter Tuning

```python
param={
    'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(regressor,param,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(x_train,y_train)
```

```python
grid. best_params_
```

```python
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
selectedmodel=DecisionTreeRegressor(criterion='friedman_mse',max_depth=4,max_features='sqrt',splitter='best')
```

```python
selectedmodel.fit(x_train,y_train)
```

```python
#visualization of the decision tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(20,10))
tree.plot_tree(selectedmodel,filled=True)
```
